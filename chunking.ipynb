```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
company_policy = """
Welcome to the Liflu platform. We are dedicated to connecting creators with top brands. 

Payment Terms: All creators will be paid within Net-30 days of campaign completion. If a brand fails to pay, Liflu will cover the cost up to $5,000.

Refunds: Brands can request a full refund within 14 days if the creator misses the delivery deadline by more than 48 hours.
"""
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    separators = ["\n\n", "\n", " ", ""]
)
```

```python
chunks = text_splitter.create_documents([company_policy])
```

```python
print(f"Total chunks created: {len(chunks)}\n")
```

```python
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i + 1} ---")
    print(chunk.page_content)
    print(f"(Length: {len(chunk.page_content)} chars)\n")
```
