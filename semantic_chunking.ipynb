```python
load_dotenv()
```

```python
embed_model = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5")
```

```python
semantic_splitter = SemanticChunker(
    embed_model,
    breakpoint_threshold_type="percentile"
)
```

```python
weird_document = """
The Roman Empire was one of the greatest civilizations in human history. 
Julius Caesar played a critical role in its rise to power. 
Legions of soldiers conquered vast territories across Europe.

Black holes are regions of spacetime where gravity is so strong that nothing can escape. 
Event horizons mark the boundary of no return. 
Stephen Hawking contributed massively to our understanding of this cosmic phenomenon.

To bake the perfect chocolate chip cookie, you must brown the butter first. 
Let the dough rest in the fridge for at least 24 hours to develop the flavor. 
Bake at 350 degrees until the edges are golden.
"""
```

```python
chunks = semantic_splitter.create_documents([weird_document])
```

```python
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i + 1} ---")
    print(chunk.page_content.strip())
    print("\n")
```
